```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.drop_duplicates(inplace=True)
```

```python
date_cols = [
    'Date Of Account Opening',
    'Last Transaction Date',
    'Transaction Date',
    'Approval/Rejection Date',
    'Payment Due Date',
    'Last Credit Card Payment Date',
    'Feedback Date',
    'Resolution Date'
]

for col in date_cols:
    data[col] = pd.to_datetime(data[col], errors='coerce')
```

```python
data['Full_Name']=data['First Name'] + " " + data['Last Name']
```

```python
data['Age_Group']=pd.cut(
    data['Age'],
    bins=[18,30,45,60,75],
    labels=['Young','Adult','Middel Age','Senior']
)
```

```python
data['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='Account Type', data=data)
plt.show()
```

```python
data['Loan Status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
data['Account Balance'].describe()
```

```python
sns.histplot(data['Transaction Amount'], bins=30)
plt.show()
```

```python
data['Month'] = data['Transaction Date'].dt.month

monthly = data.groupby('Month')['Transaction Amount'].sum()

monthly.plot(kind='line', marker='o',color='red')
plt.title("Monthly Transaction Trend")
plt.show()
```
